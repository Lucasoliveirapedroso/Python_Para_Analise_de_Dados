```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# L o arquivo Excel, "" tabela especfica
df = pd.read_excel("C:/Users/Lucas/OneDrive/Documentos/ibge.xlsx", 'Tab 1.1.38')

# Definir os nomes das colunas
colNm = [
    'Regiao',
    'MedioTotal',
    'MedioUtilizaram',
    'MedioNaoUtilizaram',
    'MedianoTotal',
    'MedianoUtilizaram',
    'MedianoNaoUtilizaram',
]
df.columns = colNm # Atribuir os nomes das colunas ao DataFrame
df = df.drop([0, 1, 2, 3, 46, 47]) # Remover as linhas com ndices 0, 1, 2, 3, 46 e 47
df = df.set_index(df['Regiao']) # define coluna 'Regiao' como ndice 
df = df.drop(columns=['Regiao']) # apaga a coluna regiao que no vai usuada que foi definida como indice
df = df.drop('Brasil') # Remover a linha com o rtulo 'Brasil'
df = df.sort_values('MedioTotal')# Ordenar o DataFrame por 'MedioTotal'
df.plot(rot=90) # Gera um grfico de linhas com rotao dos rtulos em 90 graus
```

```python
# Converter todas as colunas para o tipo numrico, tratando erros como valores ausentes (NaN)
df = df.apply(pd.to_numeric, errors='coerce')

# Calcular a matriz de correlao entre as colunas do DataFrame
matCorr = df.corr()  # A matriz de correlao mostra como as variveis se relacionam

# Gerar um grfico de boxplot
df.boxplot(rot=90)  # rot=90  usado para rotacionar os rtulos do eixo x em 90
```
